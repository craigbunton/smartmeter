```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV data into a DataFrame
df_summer_wd = pd.read_csv('data_2013/summary/summer_wd.csv')
df_summer_we = pd.read_csv('data_2013/summary/summer_we.csv')
df_winter_wd = pd.read_csv('data_2013/summary/winter_wd.csv')
df_winter_we = pd.read_csv('data_2013/summary/winter_we.csv')
df_summer_wd.drop('LCLid', axis=1, inplace=True)
df_summer_we.drop('LCLid', axis=1, inplace=True)
df_winter_wd.drop('LCLid', axis=1, inplace=True)
df_winter_we.drop('LCLid', axis=1, inplace=True)
# Calculate the mean KWH for each hour across all rows
mean_summer_wd_kwh_by_hour = df_summer_wd.mean()
mean_summer_we_kwh_by_hour = df_summer_we.mean()
mean_winter_wd_kwh_by_hour = df_winter_wd.mean()
mean_winter_we_kwh_by_hour = df_winter_we.mean()

# Create a line plot
plt.figure(figsize=(12, 6))
plt1 = plt.plot(mean_summer_wd_kwh_by_hour.index, mean_summer_wd_kwh_by_hour.values, color="red")
plt2 = plt.plot(mean_summer_we_kwh_by_hour.index, mean_summer_we_kwh_by_hour.values, color="orange")
plt3 = plt.plot(mean_winter_wd_kwh_by_hour.index, mean_winter_wd_kwh_by_hour.values, color="darkblue")
plt4 = plt.plot(mean_winter_we_kwh_by_hour.index, mean_winter_we_kwh_by_hour.values, color="lightblue")

plt.title('Mean KWH by Hour')
plt.xlabel('Hour')
plt.ylabel('Mean KWH')
plt.legend(['Summer-weekdays', 'Summer-weekends', 'Winter-weekdays', 'Winter-weekends'])
plt.grid(color='grey', linestyle='-', linewidth=0.25)
plt.show()
```
