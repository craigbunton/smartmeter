```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('data_2013/summary_2013.csv')

df.head()
```

```python
month_list = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
df['annual'] = df[month_list].mean(axis=1)
annual_df = df[['LCLid', 'annual']]
plt.hist(annual_df["annual"], bins=300)
plt.xlim(0, 1.0)
plt.show()
```
